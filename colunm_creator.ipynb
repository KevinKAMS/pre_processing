# Criação de novas colunas

Agora utilizamos nossa base de dados limpa para criar algumas colunas que possam nos gerar informações.
- A coluna new_acess identificará se é um primeiro acesso do dispositivo, identificado pelo device_age == 0.0
- A coluna ato_from_emulator vai checar que o ataque foi feito a partir de um emulador
- A coluna ato_from_root vai checar se o ataque foi feito a partir de um dispositivo com root
- A coluna old_account vai checar se aquela conta é antiga (> 1 ano)
- A coluna account_in_new_device vai checar se é o acesso de uma conta em um dispositivo diferente do primeiro dispositivo

```python
import pandas as pd
import numpy as np
logins = pd.read_parquet('clean_logins.parquet')
```

# Coluna new_acess

```python
logins['new_acess'] = np.where(logins['device_age_ms'] <= 0.001, True, False)
logins[logins.new_acess == True].head(20)
```

# Coluna ato_from_emulator

```python
logins['ato_from_emulator'] = np.where((logins['ato'] == True) & (logins['is_emulator'] == True), True, False)
logins[logins.ato_from_emulator == True].head(10)
```

# Coluna ato_from_root

```python
logins['ato_from_root'] = np.where((logins['ato'] == True) & (logins['probable_root'] == True), True, False)
logins[logins.ato_from_root == True].head(10)
```

# Coluna old_account

```python
logins['old_account'] = np.where(logins['device_age_ms'] >= 3.154e+10, True, False)
logins[logins.old_account == True].head(20)
```

# Coluna account_in_new_device

```python
#Iterar pela base de dados, para cada conta, salvar no dicionario o valor do device
#Caso ache a conta novamente iterand, checar se o valor do device é o mesmo do dicionario
#Se o valor for diferente, True, se for o mesmo ou não tiver valor, False

device_for_account = {}


def devices_and_accounts(account, device):
    if account in device_for_account:
        if device_for_account[account] == device:
            return False
        else:
            return True
    else:
        device_for_account[account] = device
        return False


logins['account_in_new_device'] = logins.apply (lambda row: devices_and_accounts(row.account_id, row.device_id), axis=1)
#logins['old_account'] = np.where(devices_and_accounts(logins['account_id'],logins['device_id']), True, False)
```
